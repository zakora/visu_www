```python
import pandas as pd
```

```python
data_rel_path = "../data/"
df = pd.read_csv(data_rel_path + "from_visualizing/2013_rankings_for_report.csv", sep=";", )
all_data = df.values
df.head()
```

```python
ct = pd.read_csv(data_rel_path + "from_wikipedia/country_codes_iso3.csv", sep="     ", header=None, names=["ISO", "Country"])
ct.head()
```

```python
full_data = pd.merge(ct, df, on="Country")
full_data.head()
```

```python
country_iso_composite = full_data[["ISO", "Index Composite"]]
iso = full_data["ISO"]
indexcomp = full_data["Index Composite"]
iso_indexcomp = {a: str(b) for a, b in zip(iso, indexcomp)}
iso_indexcomp
```

```python
import json
with open("iso_index.json", "w") as f :
    f.write(json.dumps(iso_indexcomp))
```
